```python
plt.figure(figsize=(20, 14))

for index, curr_feature  in enumerate(utils.givenFeatures):
    plt.subplot(2,2,index+1)
    plt.hist(
        df_4_genre[curr_feature],
        density=True,
        stacked=False,
        label=df_4_genre.index)
    plt.title(curr_feature)

plt.legend()
plt.savefig("../Figures/task2_distributions_stacked.png")
```
